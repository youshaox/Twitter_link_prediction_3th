# Decision tree

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV
```

```python
# 1. load the data
data_dir = "../../data/test_data/"
df = pd.read_csv(data_dir + "ad.data", header=None)
```

```python
# 3279 rows
# 1559 columns
explantory_variable_columns = set(df.columns.values)
response_variable_coulmn = df[len(df.columns.values)-1]

explantory_variable_columns.remove(len(df.columns.values)-1)

y = [1 if e=="ad." else 0 for e in response_variable_coulmn]

# 相当于用list作为一个index
X =df[list(explantory_variable_columns)]

# 用negative value来代替
X.replace(to_replace=' *\?', value=-1, regex=True, inplace=True)
```

```python
# splitting the training data and the test data set
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
pipline = Pipeline([('clf',DecisionTreeClassifier(criterion='entropy'))])
parameters = {
    'clf__max_depth' : (150, 155, 160),
    'clf__min_samples_split' : (1, 2, 3),
    'clf__min_samples_leaf' : (1, 2, 3)
}
```

```python
grid_search = GridSearchCV(pipline, parameters, n_jobs = -1, verbose = 1, scoring = 'f1')
grid_search.fit(X_train, y_train)
```
